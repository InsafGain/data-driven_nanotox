```python
!pip install rdkit
!pip install pubchempy
!pip install py3Dmol
!pip install pymatgen
!pip install boto3
from rdkit import Chem
from rdkit.Chem import Draw
import py3Dmol
import pandas as pd
from rdkit.Chem import AllChem
import pymatgen.core as pmg
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.phase_diagram import PDPlotter, PhaseDiagram
from pymatgen.io.cif import CifParser
%matplotlib inline

url = 'https://drive.google.com/uc?id=1q2zjGbjNQLphJ5L-v6-Hgjy_vnNZVEvN'
df_cleaned_modified = pd.read_csv(url)
df_cleaned_modified
df_cleaned_modified = df_cleaned_modified.drop('Unnamed: 0', axis=1)
```

```python
#Построение изображения частицы ZnO в газовой фазе
ZnO = Chem.MolFromSmiles('[Zn]O')
img = Draw.MolToImage(ZnO)
img.show()
block = Chem.MolToMolBlock(ZnO)
view = py3Dmol.view(width=400, height=400)
view.addModel(block, 'ZnO')
view.setStyle({'stick': {}, 'sphere': {}})
view.zoomTo()
view.show()
```

```python
#Построение изображения частицы Mn2O3 в газовой фазе
smiles = 'O=[Mn]O[Mn]=O'
Mn2O3 = Chem.MolFromSmiles(smiles)
img = Draw.MolToImage(Mn2O3)
img.show()
block = Chem.MolToMolBlock(Mn2O3)
view = py3Dmol.view(width=400, height=400)
view.addModel(block, 'Mn2O3')
view.setStyle({'stick': {}, 'sphere': {}})
view.zoomTo()
view.show()
```

```python
#Посмтроение фазовой диаграммы Mn-O в координатах Энергии образования/атом - мольная доля атома
a = MPRester('WlE5j5h5PnQJkdCb8wNrOkorAKYNQ42f')
entries = a.get_entries_in_chemsys(["Mn", "O"])
pd = PhaseDiagram(entries)
plotter = PDPlotter(pd, show_unstable=0.2, backend="matplotlib")
plotter.show()
```

```python
#Посмтроение фазовой диаграммы Fe-O в координатах Энергии образования/атом - мольная доля атома
entries_2 = a.get_entries_in_chemsys(["Fe", "O"])
pd_2 = PhaseDiagram(entries_2)
plotter_2 = PDPlotter(pd_2, show_unstable=0.2, backend="matplotlib")
plotter_2.show()
```
