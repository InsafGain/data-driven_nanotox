```python
import pandas as pd
url = 'https://drive.google.com/uc?id=15tY5TKOrHb-S5tt3fiQlvLyTWGgxk9Vl'
df1 = pd.read_csv(url)
df1
```

```python
url2 = 'https://drive.google.com/uc?id=1DNdX5i-GA2LjEUJCTJvwbBgCuJ32sE-j'
df2 = pd.read_csv(url2, sep = ';')
df2
df2['class'] = df2['class'].replace({'nonToxic': 0, 'Toxic': 1})
```

```python
url3 = 'https://drive.google.com/uc?id=1tP7vOkEC_4Jt2fGNo4snaL6oSmwd2p_Y'
df3 = pd.read_csv(url3, sep = ';')
df3
df3['class'] = df3['class'].replace({'Nontoxic': 0, 'Toxic': 1})
```

```python
#считали готовые датасеты, готовимся к их объединению и очистке
concatenated_df = pd.concat([df1, df2, df3], axis=1, join='outer')
concatenated_df
```

```python
#Данный способ объединения датасетов нам не подходит, используем другой подход
columns_df1 = set(df1.columns)
columns_df2 = set(df2.columns)
columns_df3 = set(df3.columns)

# Находим объединение названий колонок
unique_columns_3 = columns_df1.union(columns_df3).union(columns_df2)

# Преобразуем в список
unique_columns_list_3 = list(unique_columns_3)
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Оставляем только колонки из первого DataFrame
result_df = combined_df[unique_columns_list_3]

print(result_df)

duplicate_count = result_df.duplicated().sum()
print(f"Количество дубликатов: {duplicate_count}")

duplicate_count_df1 = df1.duplicated().sum()
print(f"Количество дубликатов: {duplicate_count_df1}")
```

```python
#датасеты содержат множество дубликатов, но прежде чем избавиться от них, избавимся от строчек, содержащих пропущенные значения
RED = '\033[91m'
BLUE = '\033[94m'
RESET = '\033[0m'

for col in result_df.columns:
    num_unique_values = result_df[col].nunique()
    if pd.api.types.is_numeric_dtype(result_df[col]):
        dtype = f"{BLUE}numerical{RESET}"
    else:
        dtype = f"{RED}categorical{RESET}"


    print(f"Column '{col}' has {num_unique_values} unique {dtype} items")
```

```python
df_cleaned = result_df.dropna(axis=1)

# Выводим очищенный DataFrame
print("\nDataFrame после удаления колонок с пропущенными значениями:")
print(df_cleaned)
```

```python
#Проверяем наличие пропущенных значений
data_types = df_cleaned.dtypes
print("Типы данных в столбцах:")
print(data_types)
missing_values = df_cleaned.isnull().sum()
print("\nКоличество пропущенных значений в каждом столбце:")
print(missing_values)
```

```python
numerical_df = df_cleaned.select_dtypes(include=['float64', 'int64'])

for column in numerical_df:
    Q1 = numerical_df[column].quantile(0.25)
    Q3 = numerical_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    numerical_df_modified = numerical_df[(numerical_df[column] >= lower_bound) & (numerical_df[column] <= upper_bound)]

filtered_indices = numerical_df_modified.index
df_cleaned_with_categories = df_cleaned.loc[filtered_indices]

df_cleaned_with_categories
```

```python
#Получаем конечный объединенный датасет, который уже можно дальше анализировать и использовать для обучения
duplicate_count = df_cleaned_with_categories.duplicated().sum()
print(f"Количество дубликатов: {duplicate_count}")

# Удаление дубликатов
df_cleaned_modified = df_cleaned_with_categories.drop_duplicates()
df_cleaned_modified
```
